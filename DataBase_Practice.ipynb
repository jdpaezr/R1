```python
path="https://raw.githubusercontent.com/jdpaezr/R1/main/Datasets-main/datasets/satis.csv"
import pandas as pd
import numpy as np
df=pd.read_csv(path)
df.head()
```

Descripcion Variables

'Age'[Numerica]: Edad de cada cliente
'Gender'[categorica]: Genero del cliente
'Ocupation'[Categorical]: A que se dedica el cliente

```python
df.isnull().any()
```

```python
favs=['Age','Gender','Price']
ndf=df[favs]
ndf.dropna(how='all',
           axis=0,
           inplace=True)
print(ndf)
```

```python
ndf['Age']=ndf['Age'].astype(int)
ndf
```

```python
ndf.isnull().any()
```

##1. Â¿Cuantas personas hay por Age Group?

```python
ndf.head(2)
```

```python
def cust_cls(x):
  if (0<=x<=20):
    return '<=20'
  elif(20<x<=30):
      return'21-30'
  elif(30<x<=40):
    return'31-40'
  else:
    return'>40'
```

```python
new_fun= lambda tbl: cust_cls(tbl['Age'])
ndf['AgeGroup'] = df.apply(new_fun,axis=1)
ndf
```

```python
my_pivot=pd.pivot_table(ndf,
                        index='AgeGroup',
                        aggfunc='count',
                        values='Price'
                        ).rename({'Price':'Numero'},
                                 axis=1)
my_pivot.reset_index(inplace=True)
my_pivot
```

```python
import matplotlib.pyplot as plt
```

```python
explode_size=[0.03,0.04,0.04,0.03]
color_favs=['gold','yellowgreen','coral','skyblue','pink']
my_pivot['Numero'].plot(kind='pie',
                        figsize=[15,6],
                        startangle=90,
                        autopct='%1.1f%%',
                        shadow=True,
                        labels=None,
                        pctdistance=1.1,
                        explode=explode_size,
                        colors=color_favs)
plt.legend(labels=my_pivot.AgeGroup,bbox_to_anchor=[1.0,1.0])
```
