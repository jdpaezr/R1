```python
#Para simploficar Codigo se usa loc(location) y iloc(index location).

#df.loc[row_name_start: row_name_end, column_name_stat:column_name_end]
#df.iloc[row_index_start: row_index_end, column_index_stat:column_index_end]
```

```python
df.loc[:,:'GenderGroup']

#row es de principio a fin por ende se ponen unicamente el doble punto para indicarlo. La columna es de principio
# hasta gender group por eso se define.
```

```python
df.iloc[:,:4]
#Para llamar las primeras 4 columnas de la base de datos.
```
