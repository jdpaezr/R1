```python

import pandas as pd
import numpy as np
```

```python
data={'company':['Ford','Ford','VW','Audi','Audi','Ford'],
   'Price': [20,22,30,34,25,28],
   'Weight': [55,60,70,28,30,80],
   'Length':[30,35,40,40,80,pd.NaT],
   'Made': ['China','Mexico','Mexico','China',np.nan,np.nan]
   }

df = pd.DataFrame(data=data)
df
```

```python
ndf=df.copy()
ndf
#para copiar la tabla nueva generada
```

```python
ndf=ndf.dropna()
ndf
#Para eliminar filas con datos perdidos
```

```python
ndf=df.copy()
ndf
#para copiar la tabla nueva generada
```

```python
ndf.isnull().any()
#para checar que variables tienen datos perdidos
```

```python
ndf.Length.mean()
#Para sacar promedio de la columna Length
```

```python
ndf.describe()
#Datos estadistcos de la tabla
```

```python
lm=ndf.Length.median()
lm
```

```python
ndf['Length'].fillna(value=lm, inplace=True)
ndf
#ndf['Length']=ndf['Length'].fillna(value=lm)
#Para remplazar un valor perdido en la tabla con la mediana.
```

```python
mm=ndf.Made.mode()[1]
mm
```

```python
ndf['Made'].fillna(value= mm ,inplace= True)
ndf
```

```python
ndf.isnull().any()
```

```python
#What is the averge price of cars made in Mexico?
##What is the averge price of Ford cars made in Mexico?
```

```python
ndf.index
```

```python
my_pivot= pd.pivot_table(ndf,index='Made',values='Price',aggfunc=['median','mean'])
my_pivot
```

```python
print('The average price of cars in mexico is',my_pivot.iloc[1,0])
```

```python
print('The average price of cars in mexico is',my_pivot.loc['Mexico',('mean','Price')])
```

```python
my_pivot2=pd.pivot_table(ndf,index=['company','Made'],values='Price',aggfunc='median')
my_pivot2
```

```python
my_pivot2.iloc[3,0]
```

```python
my_pivot2.index
```

```python
ans= my_pivot2.loc[('Ford', 'Mexico'),'Price']
ans
```

```python
print('The average price of a ford car in mexico is',ans)
```
